**Import Library**
<br />Mengimport library yang digunakan

```python
import pandas as pd
```

**Load Dataset from Spotify API** <br />Load dataset yang diambil dari API Spotify dan menampilkannya.

```python
data_API = pd.read_csv('/content/drive/MyDrive/spotify_data.csv')
data_API
```

**Drop unused feature**</br>Menghapus fitur/kolom yang tidak digunakan.

```python
data_API.drop(['Unnamed: 0'], inplace=True, axis=1)
data_API.info()
```

**Load Dataset from Kaggle** <br />Load dataset yang diambil dari [Kaggle](https://www.kaggle.com/ektanegi/spotifydata-19212020) dan menampilkannya.

```python
data_Kaggle = pd.read_csv('/content/data.csv')
data_Kaggle
```

**Drop unused feature**</br>Menghapus fitur explicit dan year.

```python
data_Kaggle.drop(['explicit', 'year'], inplace=True, axis=1)
data_Kaggle.info()
```

**Joining the Datasets**</br>Menggabungkan dataset yang diambil dari Spotify API dan Kaggle.

```python
frames = [data_API, data_Kaggle]
result = pd.concat(frames)
```

```python
result
```

**Converting dataframe to csv**

```python
result.to_csv("datasetSpotify.csv", sep = ',')
```
