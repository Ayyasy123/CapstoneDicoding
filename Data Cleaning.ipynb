```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = "datasetSpotify.csv"
df = pd.read_csv(data_path)
df
```

```python
df_new = pd.DataFrame({
    'artist_name':df['artists'],
    'music_name':df['name'],
    'music_id':df['id'],
    'release_date':df['release_date'],
    'popularity':df['popularity'],
    'acousticness':df['acousticness'],
    'danceability':df['danceability'],
    'duration_ms':df['duration_ms'],
    'energy':df['energy'],
    'instrumentalness':df['instrumentalness'],
    'liveness':df['liveness'],
    'valence':df['valence'],
    'loudness':df['loudness'],
    'speechiness':df['speechiness'],
    'tempo':df['tempo'],
    'key':df['key'],
})
df_new['artist_name'] = df_new['artist_name'].str.title()
df_new['artist_name'] = df_new['artist_name'].replace({'\[|\]|\'|\ ': '', ',': ' '}, regex=True)
df_new.head(5)
```

```python
duplicateRowsDF = df_new[df_new.duplicated(subset=['music_id'], keep=False)]
duplicateRowsDF
```

```python
df_new = df_new.drop_duplicates(subset=['music_id'], keep='first', inplace=False)
df_new
```

```python
continuous_features = df.select_dtypes(include='float').columns.to_list()
continuous_features
```

```python
import seaborn as sns

for features in df[continuous_features]:
    sns.boxplot(df_new[features], color='blue')
    plt.show()
```

```python
Q1 = df_new.quantile(0.25)
Q3 = df_new.quantile(0.75)
IQR = Q3-Q1
df_new = df_new[~((df_new<(Q1-1.5*IQR))|(df_new>(Q3+1.5*IQR))).any(axis=1)]
```

```python
df_new.shape
```

```python
df_new.drop(df_new[df_new.popularity < 50].index, inplace=True)
```

```python
df_new
```

```python
def create_cek_data(x):
    return x['artist_name'] + ' ' + x['music_name']
df_new['cek_artist_music_name'] = df_new.apply(create_cek_data, axis=1)
```

```python
duplicateRowsDF = df_new[df_new.duplicated(subset=['cek_artist_music_name'], keep=False)]
duplicateRowsDF
```

```python
df_new[df_new['cek_artist_music_name'].eq('HarryStyles Watermelon Sugar')]
```

```python
df_new = df_new.drop_duplicates(subset=['cek_artist_music_name'], keep='first', inplace=False)
df_new
```

```python
df_clean = pd.DataFrame({
    'artist_name':df_new['artist_name'],
    'music_name':df_new['music_name'],
    'music_id':df_new['music_id'],
    'release_date':df_new['release_date'],
    'popularity':df_new['popularity'],
    'acousticness':df_new['acousticness'],
    'danceability':df_new['danceability'],
    'duration_ms':df_new['duration_ms'],
    'energy':df_new['energy'],
    'instrumentalness':df_new['instrumentalness'],
    'liveness':df_new['liveness'],
    'valence':df_new['valence'],
    'loudness':df_new['loudness'],
    'speechiness':df_new['speechiness'],
    'tempo':df_new['tempo'],
    'key':df_new['key'],
})
df_clean.head(5)
```
